# **MACHINE LEARNING**

```python
x=df.drop('Obesity',axis=1)
y=df['Obesity']
```

```python
x.shape
```

**SPLITTING**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,  x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=32)
```

**Feature Engineering**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

**model building**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_res=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
lr_acc=accuracy_score(y_test, lr_res)
print("Accuarcy LR :",lr_acc)

lr_cm=confusion_matrix(y_test, lr_res)
import seaborn as sns
sns.heatmap(lr_cm, annot=True, fmt='d')
plt.title("Heat map for confusion matrix in LR model")
plt.xlabel("predicted value")
plt.ylabel("Actual value")
plt.show()
```

```python
def obesityPrediction(Gender="Male", Age=43, Height=180, Weight=87, family_history="no", FAVC="no", FCVC=2,NCP=3, CAEC="Sometimes", SMOKE="no", CH2O=2, SCC="no", FAF=3, TUE=1 ,CALC="Sometimes", MTRANS="Bike"):
    lst=[]
    if Gender=="Female":
        lst=lst+[0]
    elif Gender=="Male":
        lst=lst+[1]

    lst=lst+[Age]
    lst=lst+[Height]
    lst=lst+[Weight]
    
    if family_history=="no":
        lst=lst+[0]
    elif family_history=="yes":
        lst=lst+[1]
        
    if FAVC=="no":
        lst=lst+[0]
    elif FAVC=="yes":
        lst=lst+[1]
        
    lst=lst+[FCVC]
    lst=lst+[NCP]
    
    CAEC=lb_obesity.transform([CAEC])
    lst=lst+list(CAEC)
    
    if SMOKE=="no":
        lst=lst+[0]
    elif SMOKE=="yes":
        lst=lst+[1]

    lst=lst+[CH2O]
    
    if SCC=="no":
        lst=lst+[0]
    elif SCC=="yes":
        lst=lst+[1]

    lst=lst+[FAF]
    lst=lst+[TUE]
    
    CALC=lb1_obesity.transform([CALC])
    lst=lst+list(CALC)

    MTRANS=lb2_obesity.transform([MTRANS])
    lst=lst+list(MTRANS)

    lst=sc.transform([lst])
    
    result=obesity_model.predict(lst)
    print(result)
    if result==0:
        return "Insufficient_Weight"
    elif result==1:
        return 'Normal_Weight'
    elif result==2:
        return 'Obesity_Type_I'
    elif result==3:
        return 'Obesity_Type_II'
    elif result==4:
        return 'Obesity_Type_III'
    elif result==5:
        return 'Overweight_Level_I'
    elif result==6:
        return 'Overweight_Level_II'
    
# ['Insufficient_Weight',
#  'Normal_Weight',
#  'Obesity_Type_I',
#  'Obesity_Type_II',
#  'Obesity_Type_III',
#  'Overweight_Level_I',
#  'Overweight_Level_II']
```
